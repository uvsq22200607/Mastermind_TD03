```python
import tkinter as tk
from tkinter import *
import tkinter.messagebox
from random import randint


def quitter():
    root.destroy()

def color(a):
    global count1
    global count2
    global count3
    global count4
    if a == 1:
        if count1%4 == 0:
            canvas1.config(bg="red")
        elif count1%4 == 1:
            canvas1.config(bg="blue")
        elif count1%4 == 2:
            canvas1.config(bg="green")
        elif count1%4 == 3:
            canvas1.config(bg="yellow")
    elif a == 2:
        if count2%4 == 0:
            canvas2.config(bg="red")
        elif count2%4 == 1:
            canvas2.config(bg="blue")
        elif count2%4 == 2:
            canvas2.config(bg="green")
        elif count2%4 == 3:
            canvas2.config(bg="yellow")
    elif a == 3:
        if count3%4 == 0:
            canvas3.config(bg="red")
        elif count3%4 == 1:
            canvas3.config(bg="blue")
        elif count3%4 == 2:
            canvas3.config(bg="green")
        elif count3%4 == 3:
            canvas3.config(bg="yellow")
    elif a == 4:
        if count4%4 == 0:
            canvas4.config(bg="red")
        elif count4%4 == 1:
            canvas4.config(bg="blue")
        elif count4%4 == 2:
            canvas4.config(bg="green")
        elif count4%4 == 3:
            canvas4.config(bg="yellow")

def clicked1(event):
    global count1
    count1 += 1

def clicked2(event):
    global count2
    count2 += 1

def clicked3(event):
    global count3
    count3 += 1

def clicked4(event):
    global count4
    count4 += 1

def validate():
    ganswer = 0
    if count1%4+1 == combination[0]:
        answer1= True
        ganswer+=1
    else:
        answer1= False
    if count2%4+1 == combination[1]:
        answer2= True
        ganswer+=1
    else:
        answer2= False
    if count3%4+1 == combination[2]:
        answer3= True
        ganswer+=1
    else:
        answer3= False
    if count4%4+1 == combination[3]:
        answer4= True
        ganswer+=1
    else:
        answer4= False
    
    lab_ga.config(text = "nombre de bonnes reponses : " + str(ganswer))
    



count1=0
count2=0
count3=0
count4=0
combination = (randint(1,4),randint(1,4),randint(1,4),randint(1,4)) 

root = tk.Tk()
root.title("MASTERMIND")

canvas1 = tk.Canvas(root, width=76, height= 75, bg="red")
canvas1.grid(row=1, column= 1)
canvas2 = tk.Canvas(root, width=75, height= 75, bg="red")
canvas2.grid(row=1, column= 2)
canvas3 = tk.Canvas(root, width=75, height= 75, bg="red")
canvas3.grid(row=1, column= 3)
canvas4 = tk.Canvas(root, width=75, height= 75, bg="red")
canvas4.grid(row=1, column= 4)


buttonQ = tk.Button(root, width=10, height=5, command=quitter,text="Quitter")
buttonQ.grid(row = 0, column= 0)


buttonColor1 = tk.Button(root, width=10, height=5, text="",command=lambda : color(1))
buttonColor1.grid(row = 2, column= 1)
buttonColor2 = tk.Button(root, width=10, height=5, text="",command=lambda : color(2))
buttonColor2.grid(row = 2, column= 2)
buttonColor3 = tk.Button(root, width=10, height=5, text="",command=lambda : color(3))
buttonColor3.grid(row = 2, column= 3)
buttonColor4 = tk.Button(root, width=10, height=5, text="",command=lambda : color(4))
buttonColor4.grid(row = 2, column= 4)

buttonV = tk.Button(root, width=10, height=5, text="Valider",command=validate)
buttonV.grid(row = 2, column= 5)

lab_ga=tk.Label(root,text="")
lab_ga.grid(column=0,row=3)

lab_C=tk.Label(root,text="bonne combinaison : " + str(combination))
lab_C.grid(column=0,row=4)

buttonColor1.bind("<Button-1>",clicked1)
buttonColor2.bind("<Button-1>",clicked2)
buttonColor3.bind("<Button-1>",clicked3)
buttonColor4.bind("<Button-1>",clicked4)


root.mainloop()
```
