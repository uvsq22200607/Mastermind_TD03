```python
import tkinter as tk
from tkinter import *
import tkinter.messagebox
from random import randint


nb_ligne = int(input("combien de lignes"))
combinaison = [randint(1,4),randint(1,4),randint(1,4),randint(1,4)] 
parameterV = 0
ligne = 0

def quitter():
    root.destroy()

def color(canvas):
    color = tableau_canvas[ligne//4][canvas].itemcget(circle, 'fill')
    if   color== 'red':
        tableau_canvas[ligne//4][canvas].itemconfig(  circle, fill='blue')
    elif color== 'blue':
        tableau_canvas[ligne//4][canvas].itemconfig(  circle, fill='green')
    elif color== 'green':
        tableau_canvas[ligne//4][canvas].itemconfig(  circle, fill='yellow')
    elif color== 'yellow':
        tableau_canvas[ligne//4][canvas].itemconfig(  circle, fill='red')
    
    
def clicked1(event):
    tableau_count[ligne//4][0] += 1

def clicked2(event):
    tableau_count[ligne//4][1] += 1

def clicked3(event):
    tableau_count[ligne//4][2] += 1

def clicked4(event):
    tableau_count[ligne//4][3] += 1


def validate():
    global ligne
    ganswer = 0
    banswer = 0
    combinaison_copy = combinaison.copy()
    
    for i in range(len(combinaison)):
        if tableau_count[ligne//4][i]%4+1 == combinaison[i]:
            ganswer+=1  

    for i in range(len(combinaison)):
        if tableau_count[ligne//4][i]%4+1 in combinaison_copy:
            combinaison_copy.remove(tableau_count[ligne//4][i]%4+1)
            banswer +=1

    tableau_label[ligne//4].config(text = "bonnes couleurs, bien placées : " + str(ganswer)  + '\n' + "bonnes couleurs, mal placées : " + str(banswer-ganswer))
    ligne+=4
    if ganswer == 4:
        tkinter.messagebox.showinfo("CONGRATULATIONS","Well played, you won!")
        root.destroy()
    if ligne == nb_ligne*4 and ganswer != 4:
        tkinter.messagebox.showinfo("YOU LOST","You'll do better next time")
        root.destroy()


root = tk.Tk()
root.title("MASTERMIND")

tableau_count=[]
tableau_canvas = []
tableau_circle = []
tableau_label = []

for i in range(nb_ligne):
    row=[]
    for f in range(4):
        row.append(0)
    tableau_count.append(row)

for j in range(1,nb_ligne+1):
    ligne_boucle = []
    for i in range(1,5):
        canvas = tk.Canvas(root, width=75, height= 75)
        canvas.grid(row=j, column= i)
        ligne_boucle.append(canvas)
    tableau_canvas.append(ligne_boucle)

for j in range(nb_ligne):
    ligne_boucle = []
    for i in range(4):
        circle = tableau_canvas[j][i].create_oval(5, 5, 70, 70, fill='red')
        ligne_boucle.append(circle)
    tableau_circle.append(ligne_boucle)

for i in range(1,nb_ligne+1):
    lab_ga=tk.Label(root,text="")
    lab_ga.grid(column=0,row=i)
    tableau_label.append(lab_ga)

buttonQ = tk.Button(root, width=10, height=5, command=quitter,text="Quitter")
buttonQ.grid(row = 99, column= 0)


buttonColor1 = tk.Button(root, width=10, height=5, text="",command=lambda : color(0))
buttonColor1.grid(row = 99, column= 1)
buttonColor2 = tk.Button(root, width=10, height=5, text="",command=lambda : color(1))
buttonColor2.grid(row = 99, column= 2)
buttonColor3 = tk.Button(root, width=10, height=5, text="",command=lambda : color(2))
buttonColor3.grid(row = 99, column= 3)
buttonColor4 = tk.Button(root, width=10, height=5, text="",command=lambda : color(3))
buttonColor4.grid(row = 99, column= 4)


buttonV = tk.Button(root, width=10, height=5, text="Valider",command=validate)
buttonV.grid(row = 99, column= 5)

buttonColor1.bind("<Button-1>",clicked1)
buttonColor2.bind("<Button-1>",clicked2)
buttonColor3.bind("<Button-1>",clicked3)
buttonColor4.bind("<Button-1>",clicked4)

root.mainloop()
```
